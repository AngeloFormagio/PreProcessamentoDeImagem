```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

def processar_imagem(caminho_imagem):
    # Abrir a imagem
    imagem = Image.open(caminho_imagem)

    # Verificar se a imagem est em escala de cinza
    if imagem.mode != 'L':
        # Se no estiver em escala de cinza, convertemos
        imagem_cinza = imagem.convert("L")
    else:
        imagem_cinza = imagem

    # Converter a imagem para binarizada (preto e branco)
    # Usaremos um limiar para a converso (exemplo: 128)
    limiar = 128
    imagem_binarizada = imagem_cinza.point(lambda p: p > limiar and 255)

    # Exibir as imagens
    fig, axs = plt.subplots(1, 3, figsize=(18, 6))

    # Exibir a imagem original
    axs[0].imshow(imagem)
    axs[0].set_title("Imagem Original")
    axs[0].axis('off')

    # Exibir a imagem em tons de cinza
    axs[1].imshow(imagem_cinza, cmap='gray')
    axs[1].set_title("Imagem em Tons de Cinza")
    axs[1].axis('off')

    # Exibir a imagem binarizada
    axs[2].imshow(imagem_binarizada, cmap='gray')
    axs[2].set_title("Imagem Binarizada")
    axs[2].axis('off')

    # Mostrar as imagens
    plt.tight_layout()
    plt.show()

# Caminho da imagem de entrada
caminho_imagem = '/content/1.jpg'  # Substitua pelo caminho da sua imagem

# Processar e exibir as imagens
processar_imagem(caminho_imagem)
```
